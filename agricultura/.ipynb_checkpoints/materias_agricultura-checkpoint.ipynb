```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('agricultural_raw_material.csv')
```

```python
df.info()
df.isnull().sum()
```

```python
df = df.replace('%', '', regex=True)
df = df.replace(',', '', regex=True)
df = df.replace('-', '', regex=True)
df = df.replace('', np.nan)
df = df.replace('MAY90', np.nan)

df = df.dropna()


lst = ['Preço da lã grossa','Preço da lã grossa % de variação','Preço da copra','Preço da copra % de variação','Preço do algodão','Preço do algodão % de variação','Preço da lã fina','Preço da lã fina % de variação','Preço da tora dura','Preço da tora dura % de variação','Preço da madeira serrada dura','Preço da madeira serrada dura % de variação','Preço do couro','Preço do couro % de variação','Preço do compensado','Preço do compensado % de variação','Preço da borracha','Preço da borracha % de variação','Preço da tora macia','Preço da tora macia % de variação','Preço da madeira serrada macia','Preço da madeira serrada macia % de variação','Preço da polpa de madeira','Preço da polpa de madeira % de variação']
df[lst] = df[lst].astype("float")
df.dtypes
df.isnull().sum()
```

```python
df.head()
```

```python
df['Mês'] = pd.to_datetime(df['Mês'].str.upper(), format='%b%y', yearfirst=False)
df = df.set_index('Mês')
```

```python
df.head()
```

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
#raw-materials list
raw_data=['Preço da lã grossa', 'Preço da copra', 'Preço do algodão', 'Preço da lã fina', 'Preço da tora dura', 'Preço da madeira serrada dura',
'Preço do couro', 'Preço do compensado', 'Preço da borracha', 'Preço da tora macia', 'Preço da madeira serrada macia', 'Preço da polpa de madeira']
#getting the correlation matrix
corrmat = df[raw_data].corr()
#setting the size of plot
fig = plt.figure(figsize = (12, 9))
#masking the upper traingle part since matrix is symmetric(repetitive) 
mask = np.triu(np.ones_like(corrmat, dtype=bool))
sns.heatmap(corrmat, vmax = .8,mask=mask, square = True, annot = True)
plt.show()
```
