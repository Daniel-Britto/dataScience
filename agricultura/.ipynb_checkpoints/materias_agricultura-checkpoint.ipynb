```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('agricultural_raw_material.csv')
```

```python
df.info()
df.isnull().sum()
```

```python
df = df.replace('%', '', regex=True)
df = df.replace(',', '', regex=True)
df = df.replace('-', '', regex=True)
df = df.replace('', np.nan)
df = df.replace('MAY90', np.nan)

df = df.dropna()


lst = ['Preço da lã grossa','Preço da lã grossa % de variação','Preço da copra','Preço da copra % de variação','Preço do algodão','Preço do algodão % de variação','Preço da lã fina','Preço da lã fina % de variação','Preço do tronco duro','Preço do tronco duro % de variação','Preço da madeira serrada dura','Preço da madeira serrada dura % de variação','Ocultar preço','Ocultar preço % de variação','Preço do compensado','Preço do compensado % de variação','Preço da borracha','Preço da borracha % de variação','Preço do tronco macio','Preço do tronco macio % de variação','Preço da madeira serrada macia','Preço da madeira serrada macia % de variação','Preço da polpa de madeira','Preço da polpa de madeira % de variação']
df[lst] = df[lst].astype("float")
df.dtypes
df.isnull().sum()
```

```python
df.head()
```

```python
df['Mês'] = pd.to_datetime(df['Mês'].str.upper(), format='%b%y', yearfirst=False)
df = df.set_index('Mês')
```
